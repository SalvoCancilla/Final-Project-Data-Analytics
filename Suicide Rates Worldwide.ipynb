```python
# RINOMINO COLONNE
df = df.rename(columns={'Unnamed: 0':'Country', 'Unnamed: 1':'Year',
                        'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease':
                            'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Both sexes)',
                            'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease.1':
                                'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Male)',
                                'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease.2':
                                    'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Female)',
                                    'Crude suicide rates (per 100 000 population)':'Crude suicide rates (Both sexes)',
                                    'Crude suicide rates (per 100 000 population).1':'Crude suicide rates (Male)',
                                    'Crude suicide rates (per 100 000 population).2':'Crude suicide rates (Female)'})
                             
```

```python
# ELIMINA IL RECORD 0 INUTILE
df=df.drop(labels=0, axis=0)
```

```python
df.info()
```

```python
# YEAR IN DATETIME
df["Year"]=pd.to_datetime(df["Year"], format="%Y")
```

```python
# PRENDE SOLO L' ANNO DELLA COLONNA YEAR, MA RESTITUISCE L' OUTPUT IN FLOAT
df['Year'] = df['Year'].dt.year
```

```python
# RIASSEGNA IL DATATYPE DATETIME ALLA COLONNA YEAR
df["Year"]=pd.to_datetime(df["Year"], format="%Y")
```

```python
df.info()
```

```python
# TOLTI INTERVALLI DI CONFIDENZA
df['Crude suicide rates (Female)'] = df['Crude suicide rates (Female)'].str.extract(r'(\d+\.\d+)')
df['Crude suicide rates (Male)'] = df['Crude suicide rates (Male)'].str.extract(r'(\d+\.\d+)')
df['Crude suicide rates (Both sexes)'] = df['Crude suicide rates (Both sexes)'].str.extract(r'(\d+\.\d+)')
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Female)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Female)'].str.extract(r'(\d+\.\d+)')
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Male)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Male)'].str.extract(r'(\d+\.\d+)')
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Both sexes)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Both sexes)'].str.extract(r'(\d+\.\d+)')
```

```python
# DATATYPE IN FLOAT
df['Crude suicide rates (Female)'] = df['Crude suicide rates (Female)'].astype(float)
df['Crude suicide rates (Male)'] = df['Crude suicide rates (Male)'].astype(float)
df['Crude suicide rates (Both sexes)'] = df['Crude suicide rates (Both sexes)'].astype(float)
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Female)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Female)'].astype(float)
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Male)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Male)'].astype(float)
df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Both sexes)'] = df['Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease (Both sexes)'].astype(float)
```
