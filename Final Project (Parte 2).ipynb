# DATASET PROBLEMI MENTALI (--- DA USARE ---)
(2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010,2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019)

```python
df=pd.read_csv('Dataset problemi mentali\Mental health Depression disorder Data.csv')
df
```

```python
df.info()
```

```python
# ELIMINO COLONNE INUTI
df=df.drop('index', axis=1)
df=df.drop('Code', axis=1)
```

```python
# DATATYPE DATETIME ALLA COLONNA YEAR

df=df[df["Year"].str.isnumeric()]
df["Year"] = df["Year"].astype(int)
df = df.loc[(df["Year"] >= 2000) & (df["Year"] <= 2019)]
df["Year"]=pd.to_datetime(df["Year"], format="%Y")
df['Year'] = df['Year'].dt.year
df.Year.unique()
```

```python
# DATATYPE FLOAT ALLE ALTRE COLONNE
df["Schizophrenia (%)"]=df["Schizophrenia (%)"].astype(float)
df["Bipolar disorder (%)"]=df["Bipolar disorder (%)"].astype(float)
df["Eating disorders (%)"]=df["Eating disorders (%)"].astype(float)
```

```python
# RINOMINO COLONNA ENTITY IN COUNTRY 
df=df.rename(columns={'Entity': 'Country'})
```

```python
df_mental=df.copy()
df_mental
```

```python
# MERGE CON I PRECEDENTI DATASET
merged_df = pd.merge(merged_df,df, on=['Year','Country'], how='left')
```
